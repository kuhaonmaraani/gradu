```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

Calculate VTEC from STEC

```python
path = 'losdf/losdf1.csv'

df = pd.read_csv(path, sep=',')

df
```

```python
df['minute'] = df['min']
df['second'] = df['sec']
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute', 'second']])
df = df[['datetime', 'pierce_alt', 'gdlatr', 'gdlonr', 'los_tec', 'tec', 'azm', 'elm', 'gdlat', 'glon', 'rec_bias']]
df['slant_f'] = 1+16*(0.53-df['elm']/180)**3
df['vtec'] = (df['los_tec'])/df['slant_f']

df.head(5)
```

```python
df1 = df[['datetime', 'tec', 'vtec']]
df1['dtec'] = df1['tec'] - df1['vtec']
df1['dtec'].describe()
```
